## What is Token Fertility?

#### Token Fertility?
> Token fertility measures the average number of tokens (subwords) a large language model (LLM) generates for each word in a given text.

**Lower Fertility = Better Efficiency**
- Reduced inference cost →  Because you pay for LLMs on a per-token basis, higher fertility (more tokens per word) results in higher API bills and computational load
- Efficient use of context window →  Because you pay for LLMs on a per-token basis, higher fertility (more tokens per word) results in higher API bills and computational load
- Lower latency → better for real-time apps like voice assistants
- Improved alignment and training → more natural linguistic boundaries
- Model Accuracy: High fertility rates frequently correlate with lower reasoning and question-answering accuracy, as the fragmented subwords make it harder for the model to establish logical patterns

## Tokenizer prepare methodology

#### Three method
| Method                            | When to Use                                                 | What It Does                                       | Result                                        |
| --------------------------------- | ----------------------------------------------------------- | -------------------------------------------------- | --------------------------------------------- |
| **BPE / SentencePiece**           | Building a **NEW** model from zero                          | Learns token patterns from your text corpus        | Custom vocabulary optimized for your language |
| **Continued BPE** (Gnani.ai)      | Adding languages to an **EXISTING** model (Nemotron, LLaMA) | Keeps base merge rules, adds new ones on top       | Best fertility, handles unknown words well    |
| **Token Expansion** (add\_tokens) | Need a **QUICK** prototype                                  | Trains separate tokenizer, copies tokens into base | Fast to build, but cannot compose new words   |


#### comparison
| Feature                  | BPE (New) | Continued BPE | add\_tokens |
| ------------------------ | --------- | ------------- | ----------- |
| Complexity               | Medium    | High          | Low         |
| Fertility (lower=better) | Excellent | Excellent     | Good        |
| Handles new words        | Yes       | Yes           | No          |
| Best for                 | New model | Production    | Prototype   |



##### BPE(Byte pair encoding) Method
BPE (Byte-Pair Encoding) is like a compression algorithm.
It starts with single characters, then repeatedly merges the most frequent pairs into new tokens.


```bash
Step 1: Start with individual characters
  Input:  ব + া + ং + ল + া
  Tokens: ["ব"] ["া"] ["ং"] ["ল"] ["া"]  -> 5 tokens

Step 2: Find most frequent pair: "ব" + "া" appears a lot
  Merge:  "ব" + "া" = "বা"
  Tokens: ["বা"] ["ং"] ["ল"] ["া"]  -> 4 tokens

Step 3: Next frequent pair: "বা" + "ং"
  Merge:  "বা" + "ং" = "বাং"
  Tokens: ["বাং"] ["ল"] ["া"]  -> 3 tokens

Step 4: Next frequent pair: "ল" + "া"
  Merge:  "ল" + "া" = "লা"
  Tokens: ["বাং"] ["লা"]  -> 2 tokens

Step 5: Final merge: "বাং" + "লা"
  Merge:  "বাং" + "লা" = "বাংলা"
  Tokens: ["বাংলা"]  -> 1 token ✓
```

### Tokenizer training with Transformers

````python


# Step 1: Install required packages
!uv add tokenizers transformers

from tokenizers import Tokenizer, models, pre_tokenizers, trainers

# Step 2: Create a BPE tokenizer
tokenizer = Tokenizer(models.BPE())

# Step 3: Set pre-tokenizer (splits text by whitespace)
tokenizer.pre_tokenizer = pre_tokenizers.Whitespace()

# Step 4: Create trainer
# vocab_size = how many unique tokens you want
# min_frequency = a pair must appear at least this many times to become a token
trainer = trainers.BpeTrainer(
    vocab_size=1000,              # Start small for demo
    min_frequency=2,              # Minimum 2 occurrences
    special_tokens=["<pad>", "<unk>", "<s>", "</s>"]
)

# Step 5: Prepare training data
corpus_lines = [
    "আমি বাংলায় কথা বলি",
    "বাংলা আমার মাতৃভাষা",
    "আমি বাংলা ভাষায় গর্বিত",
    "বাংলাদেশ আমার দেশ",
    "আমি বাংলায় কথা বলতে পছন্দ করি",
    "বাংলা ভাষা সুন্দর",
    "আমি বাংলা শিখছি"
]

with open("sample_bangla.txt", "w", encoding="utf-8") as f:
    for line in corpus_lines:
        f.write(line + "\n")

# Step 6: Train the tokenizer
print("[Training BPE tokenizer on sample corpus...]")
tokenizer.train(["sample_bangla.txt"], trainer)

# Step 7: Save the tokenizer
tokenizer.save("bangla_bpe_tokenizer.json")
print("[Tokenizer saved to bangla_bpe_tokenizer.json]")

# Step 8: Test the tokenizer
test_sentences = [
    "আমি বাংলায় কথা বলি",
    "বাংলাদেশ সুন্দর",
    "আমি বাংলা শিখছি"
]

for sentence in test_sentences:
    output = tokenizer.encode(sentence)
    print(f"Input:  {sentence}")
    print(f"Tokens: {output.tokens}")
    print(f"Count:  {len(output.tokens)} tokens\n")

# Step 9: Show the vocabulary
vocab = tokenizer.get_vocab()
sorted_vocab = sorted(vocab.items(), key=lambda x: x[1])
for token, id in sorted_vocab[:20]:
    print(f"  {id:4d}: '{token}'")

print(f"\nTotal vocabulary size: {len(vocab)}")
````
